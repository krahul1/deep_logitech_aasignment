```python
from bs4 import BeautifulSoup as soup
import requests
```

```python
from datetime import date
today = date.today()

d = today.strftime("%m-%d-%y")
print("date-", d)
```

```python
time_url = "https://time.com/".format(d)
time_url
```

```python
html = requests.get(time_url)
```

```python
bsobj = soup(html.content)
bsobj
```

```python
for link in bsobj.findAll("h3"):
    print("Headline:{}".format(link.text))
```

```python
for news in bsobj.findAll('article',{'class':'template-article'}):
    print(news.text.strip())
```
